```python
import pandas as pd


df = pd.read_csv("overall_80K.csv")
df
```

```python
df = df.drop_duplicates(subset=["target", "source"])
df
```

```python
df = df.drop(df[df.target == df.source].index)
df
```

```python
import string


def tokenize(s):
    punct = string.punctuation
    for el in punct:
        s = s.replace(el, "")
    return s.split()
```

```python
tokens = tokenize(df.loc[43]["target"])
score_list = [morph.parse(word) for word in tokens]
[el[0] for el in score_list]
```

```python
import pymorphy2
from tqdm import tqdm
import numpy as np


def get_mean_score(s, morph):
    tokens = tokenize(s)
    score_list = []
    for word in tokens:
        parsed = morph.parse(word)[0]
        method_names = [a[0].__class__.__name__ for a in parsed.methods_stack]
        not_fakes = ["Fake" not in a and "Un" not in a for a in method_names]
        el = all(not_fakes)
        score_list.append(el)
    return np.mean(score_list)


morph = pymorphy2.MorphAnalyzer()
target_scores = dict()
source_scores = dict()
for i in tqdm(df.index):
    target_scores[i] = get_mean_score(df.loc[i]["target"], morph)
    source_scores[i] = get_mean_score(df.loc[i]["source"], morph)
```

```python
c = 0
for i in df.index:
    if target_scores[i] > 0.5 and source_scores[i] < 0.5:
        print(i, df.loc[i]["target"])
        c += 1
c
```
